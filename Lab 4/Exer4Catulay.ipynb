<h4>Task-2: </h4>
Mr.Jo Jo found that “tweetid, created-at and username” columns are sensitive thus decided to mask the values from those columns. He created  new columns “new-tweet-id”,”created-at4” to store the masked “tweet-ids” and masked “created-at” values. He decided to use “username” column to store masked usernames. 

Now, analyze the following output figure to do essential mask operations.


```python
# import library 
# datetime and calendar is a default library
import datetime
import calendar

# step 1: for the new-tweet-id

# get the tweet_id data
tweet_id = df['tweet_id']
# define empty array
new_tweet_id = []
# iterate on tweet_id
for item in tweet_id:
    # masked data formula:
    # 1. get the first 5 digits
    # 2. get the next 8 digits then reversed it
    # 3. get the last 5 digits
    # final = last 5 digits + reversed 8 digits + last 5 digits 
    first = (str(item)[:5])
    second = (str(item)[5:13])
    reversed_second = int(str(second)[::-1])
    third = (str(item)[-5:])
    mask_value = str(third) + str(reversed_second) + str(first)
    # append result
    new_tweet_id.append(mask_value) 

# get the created_at data
date_id = df['created_at']
# define empty array
date_array = []
# iterate on created_at
for item1 in date_id:
    # make a final date array
    dates = item1.split()
    date_only = dates.pop(0).replace("-", " ").split()
    for dates in date_only:
        # check for the year then minus 1 
        if len(dates) == 4:
            index = date_only.index(dates)
            year = int(dates) - 1
            date_only[index] = str(year)
    # append final date array with new year data
    date_array.append(date_only)

# function for finding the equivalent date
def findDay(date):
    born = datetime.datetime.strptime(date, '%Y %m %d').weekday()
    return (calendar.day_name[born])

# define empty array
day =[]
# check the day equivalent of the date
for item2 in date_array:
    days = ' '.join(item2)
    format_date = '{0}'.format(days)
    # call the function then append result
    day.append(findDay(format_date))
    
# define empty array
month = []
# check the month equivalent of the date
for item3 in date_array:
    # convert to int
    res = list(map(int, item3))
    for item4 in res:  
        # check if the date is month
        if item4 <= 12:
            # find month then append
            res = calendar.month_name[item4]
            month.append(res)
            
# define empty array
day_num = []       
# mask the day to 16
for item5 in month:
    day_num.append('16')
    
# format the value 
dates_formatted = list(zip(day, month, day_num))

# append the final result to created_at4
created_at4 = []
for item6 in dates_formatted:
    res_format = ' '.join(item6).replace(" ", "/")
    created_at4.append(res_format)
    

# mask username to 'yyy'
username = df['username']
username_id = []
for user in username:
    username_id.append("yyy")

# define the dataframe
dataframe = { 'new-tweet-id': new_tweet_id,
        'created-at4': created_at4,
        'username' :username_id
}

# add the column to the dataframe, display
tweet_dataset = pd.DataFrame(dataframe, columns = ['new-tweet-id', 'created-at4', 'username']) 
tweet_dataset
```

<h5> Check for Negative values in numerical data types in dataset </h5>

```python
def check_negative():
    # check negative value in UnitPrice, Quantity, CustomerID
    unit_price = dataset['UnitPrice']
    unit_price_res = (unit_price < 0).values.any() 
    quantity = dataset['Quantity']
    quantity_res = (quantity < 0).values.any()
    customer_id = dataset['CustomerID']
    customer_id_res = (customer_id < 0).values.any()
    if (unit_price_res == False) and (quantity_res  == False) and (customer_id_res  == False):
        return False
    else:
        return True
    
print("before cleaning: any negatives?: ",check_negative())
```

<h5> Get the Absolute Value of the negatives</h5>

```python
def clean_negative():
    dataset['UnitPrice'] = abs(dataset['UnitPrice'])
    dataset['Quantity'] = abs(dataset['Quantity'])
    dataset['CustomerID'] = abs(dataset['CustomerID'])
    
clean_negative()
check_negative()
print("after cleaning: any negatives?: ",check_negative())
```

<h4>Task-3: </h4>
Amazon wants to find the best customer and likes to reward 50% discount on his next purchase. Find the Customer with the highest amount of total purchase amount. The output should the one as below:

```python
def customer_purchase_amounts():
    # get the total price by the product of UnitPrice and Quantity
    total_price = np.array(dataset['UnitPrice'].multiply(dataset['Quantity'], axis="index"))
    # get the customer id convert to int val
    customer_id = dataset['CustomerID']
    customer_id_int_array = customer_id.astype(int)

    # series the customer id and total price
    total_amount = pd.Series(total_price, index = customer_id_int_array)

    # get all the customer id without duplicates
    count_freq = customer_id_int_array.value_counts().index.tolist()

    # sort by customer id
    sorted_item = []
    for item in count_freq:
        result = np.array(total_amount.get(key = item))
        sorted_item.append(result)
    sorted_item

    # get the sum of the sorted values
    sum_price = []
    for item1 in sorted_item:
        x = item1.size
        if x == 1:
            sum_price.append(item1.tolist())
        else:
            res = sum(item1)
            sum_price.append(res)
    
    # return the customer id and total price final array
    customer_sumPrice = pd.Series(sum_price, index = count_freq)
    return customer_sumPrice

# call the function
res = customer_purchase_amounts()
# get the id and value of max
customer_amount = res.max()
customer_id = res.idxmax()
print("Customer ID: {0} is the best customer since he has spent a lot on our business. His total purchase amount is {1} pesos. \n50% discount will be given on his next purchase.".format(customer_id, customer_amount))
    
```

<h4>Task-4: </h4>
Amazon wants to find out the most and the least bought items. It likes to increase the sale of least bought item by bundling it with the most bought item and ready to give special 10% discount for this bundle. The output should be the one as below.

```python
# get the data of StockCode
stock_code = dataset['StockCode']

# series the StockCode and Quantity
total_sale = pd.Series(np.array(dataset['Quantity']), index = stock_code)

# get all the StockCode without duplicates
count_sales = stock_code.value_counts().index.tolist()

# sort by StockCode
sorted_item_stock = []
for item in count_sales:
    result = np.array(total_sale.get(key = item))
    sorted_item_stock.append(result)
    
# get the sum of sales of StockCode
sum_price_quantity = []
for item1 in sorted_item_stock:
    x = item1.size
    if x == 1:
        sum_price_quantity.append(item1.tolist())
    else:
        res = sum(item1)
        sum_price_quantity.append(res)
        
# series the overall sum and count_sales
stock_quantity = pd.Series(sum_price_quantity, index = count_sales)

# get the index and value of highest sales stockcode
highest_value = stock_quantity.max()
highest_index = stock_quantity.idxmax()

# get the index and value of lowest sales stockcode
lowest_value = stock_quantity.nsmallest(keep='all')
sort_low = lowest_value.index.tolist()
# sort the lowest alphabetically display the first value
string_ints = [str(int) for int in sort_low]
lowest_index = sorted(string_ints)[0]
lowest_value = np.array(stock_quantity.get(key = int(lowest_index)))

# display
print("Most bought item is StockCode {0} with {1} in purchased quantity. \nLeast bought item is StockCode {2} with {3} in purchased quantity overall. These two items can be bundeled and provide 10%\ndiscount on this bundle to increase the sale of StockCode {2}"
     .format(highest_index, highest_value, lowest_index, lowest_value))
```
